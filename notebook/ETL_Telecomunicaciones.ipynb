Convertir a mayusculas la columna Provincia y reemplazar Capital Federal por CABA para que nos queden los mismos valores en todos los dataframe

```python
def convertir_a_mayusculas_y_reemplazar(lista_dfs):
    """
    Convierte la columna 'Provincias' a mayúsculas y reemplaza 'CAPITAL FEDERAL' por 'CABA'
    en todos los dataframes de una lista.

    Args:
        lista_dfs (list): Una lista de dataframes de Pandas.
    """

    for df, _ in lista_dfs:
        df['Provincia'] = df['Provincia'].str.upper()
        df['Provincia'] = df['Provincia'].replace('CAPITAL FEDERAL', 'CABA')

dataframes = [
    (df_velocidad,  'Accesos según velocidad por Provincia para cada trimestre'),
    (df_accesos_RangoVelocidad, 'Accesos segun rango de velocidad por provincia para cada trimestre'),
    (df_velocidad_xProvincia,  'Velocidad media de bajada por provincia'),
    (df_accesos_xTecnologia, 'Accesos según tecnología por Provincia para cada trimestre'),
    (df_Dial_BAf, 'Dial-BAF'),
    (df_penetracion_poblacion, 'Penetración de Internet por cada 100 habitantes'),
    (df_penetracion_hogares, 'Penetración de Internet por cada 100 hogares'),
]
convertir_a_mayusculas_y_reemplazar(dataframes)

for df,name in dataframes:
    print(f'**DATAFRAME** {name}')
    print('------------------------')
    print(df.head())
    print('------------------------\n')
```

Eliminar el asterisco de las columnas Año y Trimestre del dataframe df_accesos_xTecnologia

```python
import re

def limpiar_y_convertir_a_entero(valor):
    """Limpia y convierte un valor a entero, manejando diferentes tipos de datos.

    Args:
        valor: El valor a limpiar y convertir.

    Returns:
        El valor convertido a entero si es posible, de lo contrario, devuelve None o un valor por defecto.
    """

    if isinstance(valor, str):
        # Si es una cadena, eliminar caracteres no numéricos y convertir
        valor_limpio = re.sub(r'\D', '', str(valor))
        if valor_limpio:
            return int(valor_limpio)
        else:
            return None  # O cualquier otro valor por defecto
    elif isinstance(valor, int):
        # Si ya es un entero, devolverlo
        return valor
    else:
        # Si es otro tipo de dato, manejar la excepción o devolver un valor por defecto
        print(f"Tipo de dato no soportado: {type(valor)}")
        return None

# Aplicando la función a las columnas
df_accesos_xTecnologia['Año'] = df_accesos_xTecnologia['Año'].apply(limpiar_y_convertir_a_entero)
df_accesos_xTecnologia['Trimestre'] = df_accesos_xTecnologia['Trimestre'].apply(limpiar_y_convertir_a_entero)
```

```python
def visualizar_registros_invalidos(df):
    """
    Visualiza los registros donde Año o Trimestre son no numéricos o cero en todos los DataFrames de una lista.
    Maneja DataFrames que no contienen las columnas especificadas.

    Args:
        lista_dfs (list): Una lista de DataFrames de Pandas.
    """

    # Crear una máscara booleana para identificar registros no numéricos o cero
    mask_no_numerico_anio = pd.to_numeric(df['Año'], errors='coerce').isnull() | (df['Año'] == 0)
    mask_no_numerico_trimestre = pd.to_numeric(df['Trimestre'], errors='coerce').isnull() | (df['Trimestre'] == 0)

    # Combinar las máscaras y filtrar el DataFrame
    registros_invalidos = df[mask_no_numerico_anio | mask_no_numerico_trimestre]

    if not registros_invalidos.empty:
        print(registros_invalidos)
    else:
        print("No se encontraron registros con valores inválidos.")

visualizar_registros_invalidos(df_accesos_xTecnologia)
```
